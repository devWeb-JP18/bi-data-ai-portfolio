**Contexto**

Seu time no app de música quer aumentar o engajamento dos usuários, ou seja, fazer com que eles escutem mais músicas por sessão.

Precisamos aumentar o número de músicas ouvidas por sessão porque isso está diretamente ligado ao tempo de uso, retenção, monetização e valor percebido pelos usuários na plataforma

**Problema**

Ao observar os dados, o time notou que usuários que escutam faixas longas (mais de 10 minutos) escutam menos músicas por sessão — afinal, músicas longas tomam mais tempo.

**Dúvida**

Vale a pena continuar promovendo músicas longas nas recomendações e playlists? Ou devemos priorizar faixas mais curtas para aumentar o consumo?

```python
import pandas as pd
import numpy as np
```

```python
path = r"C:\Users\jpkun\OneDrive\Documentos\Trilha_Python-Data-AI\pratica\desafios-casos-reais\spotify\spotify_tracks.csv"
df = pd.read_csv(path)
df
```

```python
#Convertendo de ms para min
df['duration_min'] = df['duration_ms']/60000
df
```

```python
df_music_10_min = df.loc[df['duration_min']>10]
df_music_10_min
size_10_min = df_music_10_min.shape[0]
size_10_min
```

```python
size_df_total = df.shape[0]
size_df_total
```

```python
percent_10_min_of_total = (size_10_min/size_df_total)*100

percent_10_min_of_total
```

**Resposta**

```python
print(f"As músicas com + 10 min representam {percent_10_min_of_total:.2f}% do total de tracks analisadas")
```
