***2 - Which products generated the highest total revenue?***

```python
#1 Extract the relevant infos
#2 Clean the dataset
#3 calculate the revenue
#4 Group by the dataset
#5 sort by revenue
#6 format to Brazilian currrency


#1
df_high_revenue_products = df[['StockCode', 'Description', 'Quantity','UnitPrice']]
df_high_revenue_products
```

```python
#2
df_high_revenue_products.isna().sum()

#The empty rows are 0,2% off the dataset
df_high_revenue_products = df_high_revenue_products.dropna()
df_high_revenue_products
```

```python
#3
df_high_revenue_products['Total revenue by invoice'] = df_high_revenue_products['Quantity'] * df_high_revenue_products['UnitPrice']
df_high_revenue_products.drop(columns=['Quantity','UnitPrice'], inplace = True)
df_high_revenue_products
```

```python
#4
group_high_revenue_products = df_high_revenue_products.groupby(by=
    ['StockCode','Description'])['Total revenue by invoice'].sum().reset_index()

group_high_revenue_products
```

```python
#5
group_high_revenue_products.sort_values('Total revenue by invoice', ascending=False, inplace = True)

high_revenue_products = group_high_revenue_products.head()
high_revenue_products
```

```python
#6
import locale

# Set locale to Portuguese/Brazil
locale.setlocale(locale.LC_ALL, 'pt_BR.UTF-8')

# Format using locale currency
high_revenue_products['Total revenue'] = high_revenue_products['Total revenue by invoice'].apply(lambda x: locale.currency(x, grouping=True))
high_revenue_products.drop(columns=['Total revenue by invoice'], inplace = True)

high_revenue_products
```

**Customers & Location**

***1 - Which countries (excluding the UK) have the most purchases?***

```python
#1 Extract the rleevant dataset
#2 clean the data
#3 Group by country and apply multiple aggregations
#4 Format revenue to brazilian currency


#1
df_countries = df.loc[df['Country']!='United Kingdom',['InvoiceNo','Quantity','UnitPrice','Country']]
df_countries
```

```python
# 2
df_countries.isna().sum()
#no empty rows
```

```python
#3
df_countries['Revenue by invoice'] = df_countries['Quantity']* df_countries['UnitPrice']
df_countries.drop(columns=['Quantity','UnitPrice'])
df_countries
```

```python
group_countries = df_countries.groupby(by='Country').agg({
    'InvoiceNo':'nunique',
    'Revenue by invoice':'sum'
}).reset_index()

group_countries.rename(columns={"InvoiceNo":"Unique sales","Revenue by invoice":"Total revenue"}, inplace = True)

group_countries.sort_values(by='Unique sales', ascending = False)
```

```python
#4
import locale

# Set locale to Portuguese/Brazil
locale.setlocale(locale.LC_ALL, 'pt_BR.UTF-8')

group_countries['Total Revenue R$'] = group_countries['Total revenue'].apply(lambda x: locale.currency(x, grouping=True))
group_countries

countries_with_most_purchase = group_countries.sort_values(by=['Unique sales','Total Revenue R$'], ascending = [False,False])
countries_with_most_purchase.drop(columns='Total revenue', inplace = True)
countries_with_most_purchase
```
