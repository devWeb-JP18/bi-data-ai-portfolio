**🌍 Customer & Country Insights**

***1 - Which countries (excluding the UK) have the most purchases?***

```python
#1 Extract the rleevant dataset
#2 clean the data
#3 Group by country and apply multiple aggregations
#4 Format revenue to brazilian currency


#1
df_countries = df.loc[df['Country']!='United Kingdom',['InvoiceNo','Quantity','UnitPrice','Country']]
df_countries
```

```python
# 2
df_countries.isna().sum()
#no empty rows
```

```python
#3
df_countries['Revenue by invoice'] = df_countries['Quantity']* df_countries['UnitPrice']
df_countries.drop(columns=['Quantity','UnitPrice'])
df_countries
```

```python
group_countries = df_countries.groupby(by='Country').agg({
    'InvoiceNo':'nunique',
    'Revenue by invoice':'sum'
}).reset_index()

group_countries.rename(columns={"InvoiceNo":"Unique sales","Revenue by invoice":"Total revenue"}, inplace = True)

group_countries.sort_values(by='Unique sales', ascending = False)
```

```python
#4
import locale

# Set locale to Portuguese/Brazil
locale.setlocale(locale.LC_ALL, 'pt_BR.UTF-8')

group_countries['Total Revenue R$'] = group_countries['Total revenue'].apply(lambda x: locale.currency(x, grouping=True))
group_countries

countries_with_most_purchase = group_countries.sort_values(by=['Unique sales','Total Revenue R$'], ascending = [False,False])
countries_with_most_purchase.drop(columns='Total revenue', inplace = True)
countries_with_most_purchase
```

***3 - Which customers have spent the most overall?***

```python
#1 Extract the dataset with relevant infos
#2 clean the data
#3 extract the revenue
#4 group by customer
#5 format revenue to Brazilian currency

df_customers = df[['InvoiceNo','Quantity','UnitPrice','CustomerID']]
df_customers
df
```

```python
#2
empty_customers = df_customers['CustomerID'].isnull().sum()
percent = (empty_customers / df.shape[0])*100
percent

print("The empty rows are 24,93 % off the dataset \n")

print("Let's analyze the revenue from identified and unidentified consumers.")
```

```python
#2.1 - No empty rows
df_customers_no_empty_rows = df_customers.dropna()
df_customers_no_empty_rows
```

```python
#3.1
df_customers_no_empty_rows['Revenue'] = df_customers_no_empty_rows['UnitPrice'] * df_customers_no_empty_rows['Quantity']
df_customers_no_empty_rows['CustomerID'] = df_customers_no_empty_rows['CustomerID'].astype(int)
df_customers_no_empty_rows
```

```python
#4.1
group_customers = df_customers_no_empty_rows.groupby(by='CustomerID')['Revenue'].sum().reset_index()
group_customers.sort_values('Revenue', ascending = False, inplace = True)
total_revenue_of_identified_customers = group_customers['Revenue'].sum()
total_revenue_of_identified_customers
```

```python
#5.1
import locale

# Set locale to Portuguese/Brazil
locale.setlocale(locale.LC_ALL, 'pt_BR.UTF-8')

group_customers['Total Revenue R$'] = group_customers['Revenue'].apply(lambda x: locale.currency(x, grouping=True))
group_customers.drop(columns=['Revenue'], inplace = True)
total_revenue_of_identified_customers = locale.currency(total_revenue_of_identified_customers, grouping = True)


print(f"total revenue of identified customers: {total_revenue_of_identified_customers}")
group_customers
```

```python
#2.2 with empty rows
df_customers_with_empty_rows = df_customers.fillna("Não identificado")
df_customers_with_empty_rows
```

```python
#3.2
df_customers_with_empty_rows['Revenue'] = df_customers_with_empty_rows['UnitPrice'] * df_customers_with_empty_rows['Quantity']
df_customers_with_empty_rows
```

```python
#4.2
group_customers = df_customers_with_empty_rows.groupby(by='CustomerID')['Revenue'].sum().reset_index()
group_customers.sort_values('Revenue', ascending = False, inplace = True)

total_revenue_of_unidentified_customers = group_customers['Revenue'].sum()
print(total_revenue_of_unidentified_customers)
group_customers
```

```python
#5.2
import locale

# Set locale to Portuguese/Brazil
locale.setlocale(locale.LC_ALL, 'pt_BR.UTF-8')

group_customers['Total Revenue R$'] = group_customers['Revenue'].apply(lambda x: locale.currency(x, grouping=True))
group_customers.drop(columns=['Revenue'], inplace = True)
total_revenue_of_unidentified_customers = locale.currency(total_revenue_of_unidentified_customers, grouping = True)


print(f"total revenue of unidentified customers: {total_revenue_of_unidentified_customers}")
group_customers
```
