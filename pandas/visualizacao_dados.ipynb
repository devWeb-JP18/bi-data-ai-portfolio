```python
#Criando um df temporário para visualizar melhor o plot

# Criando o DataFrame
dados = {
    'data': pd.date_range(start='2025-01-01', periods=10, freq='D'),
    'temperatura_celsius': [12, 5, 28, 32, 9, 15, 38, 4, 25, 30],
}

# Função para classificar a temperatura
def classificar_temp(temp):
    if temp <= 5:
        return "muito frio"
    elif temp <= 15:
        return "frio"
    elif temp <= 28:
        return "quente"
    else:
        return "muito quente"

# Adicionando a coluna de classificação
df = pd.DataFrame(dados)
df['classificacao'] = df['temperatura_celsius'].apply(classificar_temp)

print(df)
```

```python

df.plot(figsize=(10,10));
```

```python
#O grid coloca as linhas horizontais no gráfico
df.plot(figsize=(10,10), grid = True);
```

```python
#style = estilizar o plot,
#-o = coloca uma bolinha nas entradas
df.plot(figsize=(10,10),grid = True, style='-o');
```

```python
#linewidth = largura da linha
df.plot(figsize=(10,10), grid = True, style='-o', linewidth = '3');
```

```python
#Color = cor da linha
df.plot(kind='line',figsize=(10,10), grid = True, linewidth = 3, style='-o', color='#ff00ff');
```

```python
#Contando a quantidade de vezes para cada classificação
df['classificacao'].value_counts().plot(
    kind='bar', figsize=(10,10), color='#ff00ff',
      grid = True, rot=45);
```

```python
#Plot de pizza
df['classificacao'].value_counts().plot(
    kind='pie', autopct='%1.1f%%',shadow =True, figsize=(10,20));
```
