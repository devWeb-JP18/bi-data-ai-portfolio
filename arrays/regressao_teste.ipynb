```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Dados fictícios: horas de estudo (x) e nota (y)
x = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)  # precisa estar em formato de matriz
y = np.array([3, 5, 7, 9, 11])  # nota

# Criando e treinando o modelo
modelo = LinearRegression()
modelo.fit(x, y)

# Fazendo a previsão da reta
x_novo = np.linspace(0, 6, 100).reshape(-1, 1)  # gera valores de 0 a 6
y_previsto = modelo.predict(x_novo)

y_previsto

# Gráfico
plt.scatter(x, y, color='blue', label='Dados reais')
plt.plot(x_novo, y_previsto, color='red', label='Linha de regressão')
plt.xlabel('Horas de estudo')
plt.ylabel('Nota')
plt.title('Regressão Linear Simples')
plt.legend()
plt.grid(True)
plt.show()
```
