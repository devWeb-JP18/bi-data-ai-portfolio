**Criando uma imagem de gráficos**

```python
x = frutas_df['Frutas']
y = frutas_df['Total Vendas']
total = frutas_df
```

**Utilizando subplot**

```python
#1 = linha, 3 = colunas, 1 = posição do gráfico
plt.subplot(231)
plt.plot(x,y,label='plot')
plt.legend()
plt.title("Gráfico 1")
plt.xticks(rotation=45)

plt.subplot(232)
plt.bar(x,y,label='Gráfico de barras')
plt.legend()
plt.xticks(rotation=45)

plt.subplot(233)
plt.pie(y, labels=x)
#plt.legend(loc='upper left', bbox_to_anchor=(1, 1))  # opcional, legenda fora



plt.tight_layout() #ajusta automaticamente os espaços

plt.show()
```
